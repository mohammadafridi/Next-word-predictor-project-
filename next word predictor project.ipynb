```python
# Initializing the Tokenizer
tokenizer = Tokenizer()
```

```python
# Fitting the tokenizer on the text
tokenizer.fit_on_texts([text])
```

```python
print(tokenizer.word_index)
```

```python
# giving every unique word index
len(tokenizer.word_index)
```

```python
input_sequence = []
for sentence in text.split('\n'):
    tokenized_sentence = tokenizer.texts_to_sequences([sentence])[0]
# input and output combo like
# input                       output
# [145]                         4790
# [145 4790]                    1
# [145 4790 1]                  1020
# so on.....
    for i in range(1,len(tokenized_sentence)):
        input_sequence.append(tokenized_sentence[:i+1])
    
```

```python
# this is only input data
input_sequence
```

```python
max_len = max([len(x) for x in input_sequence])
max_len
```

```python
# zero padding 
from tensorflow.keras.preprocessing.sequence import pad_sequences
padded_input_sequence = pad_sequences(input_sequence, maxlen = max_len, padding = 'pre')
```

```python
padded_input_sequence
```

```python
# Now I separate input and output
x = padded_input_sequence[:,:-1]
```

```python
y = padded_input_sequence[:,-1]
```

```python
x.shape
```

```python
y.shape
```

```python
# one hot encoding on y
from tensorflow.keras.utils import to_categorical
y = to_categorical(y,num_classes =616)
```

```python
y.shape
```

```python
y
```

```python
# model training
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, Dense
```

```python
model = Sequential()
model.add(Embedding(616,100, input_length=20))
model.add(LSTM(150))
model.add(Dense(256,activation='relu'))
model.add(Dense(616, activation = 'softmax'))
```

```python
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
model.summary()
```
